This notebook takes mp3 files, converts them to linear16 codec and resamples them to a bit rate of  16,000. Resampling isn't recommended, but it makes it easy to batch process because you don't need to worry about varying sample rates. Last, the files are renamed to avoid any issues with the cloud.storage API. 

```python
%%bash 
cd /home/galen/Desktop/audio
for file in *.mp3;do ffmpeg -i "$file"  -f s16le -acodec pcm_s16le -ac 1 -ar 16000 "${file%.*}".raw;done
```

```python
%%bash
cd /home/galen/Desktop/audio
j=1;for i in *.raw; do mv "$i" file"$j".raw; let j=j+1;done
```
